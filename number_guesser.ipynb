```
import tensorflow as tf
import numpy as np

x = np.array([-10, 7, 5, -33])
y = np.array([10.3, 68.7, 61.8, -68.7], dtype=float)


model = tf.keras.Sequential([tf.keras.layers.Dense(units=1, input_shape=[1])])
model.compile(loss='mean_squared_error', optimizer=tf.keras.optimizers.Adam(0.1))
history = model.fit(x, y, epochs=2000, verbose=False)

while True:
  data = float(input ('Enter the value of x = '))
  data = np.reshape(data, (1,1))
  print('Prediction for y = ', model.predict(data))

```
